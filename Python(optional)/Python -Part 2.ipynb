```python
#string operations
var1="Hello" 
var2="World"
```

## 1.concatenate two strings using + in python

```python
var=var1+var2 #concatenate strings without space,for space you explictly add " " while concatenation 
print(var) 
print(var1,var2)# prints two variables with space 
```


## 2. finding out length of the string

```python

# len function takes string as argument and returns total number of characters present in the string i.e length of a string
length=len(var)
print(length)
```

## 3. string indexing and slicing

```python

#indexing starts from 0 to length-1
a=var1[1]
print(a) # here a='e' in "Hello"
```

```python
# slicing makes to slice/cut the string as per your requirement
print(var[4:7]) #starting character from 4th index and ending with 6th index i.e upto excluding 7th index
print(var[2:3]) #prints out single 2nd index character
print(var[:7]) #prints from 0th index to 6th index
print(var[2:]) #prints from 2nd index to last character i.e length-1 index character
print(var[:]) #prints out entire characters of var variable
```

## (optional) 4. string libraries

```python

variable="Python Code"
print(variable.lower()) #lower inbuilt function converts characters to lowercase
print(variable.upper())#uppercase
print(variable.capitalize()) #first letter capitalize
greet=" Hello Mike "
print(greet.rstrip()) #removes white space at  right end which is usually present in text files
print(greet.lstrip()) #removes white space at  left end
print(greet.strip()) #removes at both the end
learn="machine learning"
found=learn.find('c') #finds 'c' character index and returns it
print(found)
new_learn=learn.replace('learning','teaching')
print(new_learn)
#replace function is like a search and replace operation in a wordprocessor
print(variable.startswith('Python'))
#startswith function returns True or False based on argument passed 
#for more inbuilt functions in string library execute the below line
print(dir(variable)) #argument is string variable
```

## Lists
Lists is a kind of collection that allows to put many values in a single variable


lists are enclosed with square bracket and each values are separated by ,commas

```python
friends=['Joseph','Glenn','Sally'] #friends store 3 string variable
var=[2,'Jack',50.4,'Mike'] #having values of string,int,float 
print(friends)
print(var)

print(friends[1]) #Glenn
```

```python
# Lists are mutable whereas strings are immutable
fruit="Banana"
#fruit[2]='h'# ,Error strings are immutable
num=[3,5,7,9]
num[2]=6
# 7 is mutated/changed to 6
print(num)
```

```python
#len function returns total values in a list
print(len(friends))
# range function returns a list of numbers that range from zero to one less than the parameter
print(range(4))
# [0,1,2,3]
print(range(6))
#  [0,1,2,3,4,5]
```

## iteration of lists

```python
friends=['Joseph','Glenn','Sally']
for i in friends:
    print('Happy New Year:',i)
for i in range(len(friends)):
    print('Happy New Year',friends[i])
```
