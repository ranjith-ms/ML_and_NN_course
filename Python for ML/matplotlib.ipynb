```python
df=pd.read_csv("kc_house_data.csv") #Load the dataset from csv file
df.head()
```

## Scatter plot

```python
plt.scatter(df['sqft_living'],df['price'],1)
plt.title("bedrooms v/s price")
plt.xlabel("Sq. ft")
plt.ylabel("Price")
plt.show()
```

## plt.plot

```python
x=df.sqft_living
line=x*300
```

```python
line
```

```python
plt.plot(x,line,'r')
plt.xlim(0,16500)
plt.show()
```

## subplots

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(x, line, 'r', label='Prediction')
ax.scatter(df['sqft_living'],df.price, label='Training Data')
ax.legend(loc=2)
ax.set_xlabel('Size')
ax.set_ylabel('Price')
ax.set_title('Predicted Price vs. Size')
```

```python
plt.subplot(211)
plt.plot(x,line)
plt.ylabel("price")
plt.xlabel("SQ. ft")
plt.title("line")

plt.subplot(2,1,2)
plt.scatter(df['sqft_living'],df.price/df.price.max())
plt.ylabel("price")
plt.xlabel("SQ. ft")
plt.title("Scatter plot")
plt.show()
```
